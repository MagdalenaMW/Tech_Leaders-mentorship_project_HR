```python
df['Attrition_1,2'] = df['Attrition'].map({'Yes': 1, 'No': 0})
df['OverTime_1,2'] = df['OverTime'].map({'Yes': 1, 'No': 0})
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])


plt.figure(figsize=(20,10))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Slides

4th slide

```python
gender_counts = df["Gender"].value_counts()
bars = plt.bar(gender_counts.index, gender_counts.values, color=["dodgerblue", "palevioletred"])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1,
             str(yval), ha='center', fontsize=12)

plt.xlabel('Gender')
plt.ylabel('Number of employees')
plt.title('Gender Distribution')
plt.show()


gender_counts = df["Gender"].value_counts()
gender_percentage = gender_counts / gender_counts.sum() * 100
plt.pie(gender_percentage, labels=gender_percentage.index, colors=["dodgerblue", "palevioletred"], autopct='%1.f%%')
plt.title('Gender Distribution [%]')
plt.show()
```

```python
plt.hist(df["Age"], bins =10, color="lightblue", edgecolor='grey')

plt.xlabel("Age")
plt.ylabel("Number of employees")


mean_age = df['Age'].mean().round()
plt.axvline(mean_age, color="red", linestyle='dashed', label=f'Mean: {mean_age:.0f}')
plt.title("Age of employees")
plt.legend()
plt.show()
```

5th slide

```python
Departments = df["Department"].value_counts()

print(Departments)


plt.pie(Departments, labels=Departments.index, autopct='%1.1f%%', startangle=90, colors=("#1E90FF", "#FFA500", "#6C9E2A"))
plt.title("Departments [%]")
plt.show()


bars = plt.bar(Departments.index, Departments.values, color=("#1E90FF", "#FFA500", "#6C9E2A"))
for bar in bars:
  yval = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, yval + 1, str(yval), ha='center', fontsize=12)
plt.title("Departments")
plt.xlabel("Department")
plt.ylabel("Number of employees")
plt.show()

```

6th slide

```python
Departments_JobRole = df.groupby(["Department", "JobRole"]).size().unstack().fillna(0)

Departments_JobRole_sorted = Departments_JobRole.loc[Departments_JobRole.sum(axis=1).sort_values(ascending=False).index]

Departments_JobRole_sorted.plot(kind='bar', stacked=True, figsize=(10, 6), colormap="Set1")
plt.title("Job Roles by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

7th slide

```python
Age_leave = df. groupby("Age")["Attrition"].value_counts().unstack()
print(Age_leave)

Age_leave.plot(kind="line")
plt.title("Age versus leaving the company")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()

print(f"Younger people until to 35 years old are more prone to leave company.")
```

```python
YearsAtCompany_EmployeeNumber = df.groupby("YearsAtCompany")["EmployeeNumber"].count()
print(YearsAtCompany_EmployeeNumber)

plt.plot(YearsAtCompany_EmployeeNumber.index, YearsAtCompany_EmployeeNumber.values)
plt.title("Number of Employees by Years at Company")
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.xlabel("Years at Company")
plt.ylabel("Number of Employees")
plt.show()
```

8th slide

```python
Attrition_YearsAtCompany = df.groupby("YearsAtCompany")["Attrition"].value_counts().unstack()

delete = [39, 40]
Attrition_YearsAtCompany_exc40 = Attrition_YearsAtCompany.drop(index=delete, errors='ignore')

full_range = range(0, 38)
Attrition_YearsAtCompany_exc40 = Attrition_YearsAtCompany_exc40.reindex(full_range, fill_value=0)

Attrition_Percent = Attrition_YearsAtCompany_exc40.div(
    Attrition_YearsAtCompany_exc40.sum(axis=1), axis=0
) * 100

Attrition_Percent = Attrition_Percent[["Yes", "No"]]

Attrition_Percent.plot(
    kind="bar",
    stacked=True,
    figsize=(14, 6),
    color=("#4f86f7", "lightgray")
)

mean_attrition_yes = Attrition_Percent["Yes"].mean()

plt.axhline(
    mean_attrition_yes,
    color='red',
    linestyle='dashed',
    linewidth=2,
    label=f'Mean Attrition Yes: {mean_attrition_yes:.2f}%'
)

plt.title("Attrition by Years at Company (%)")
plt.xlabel("Years at Company")
plt.ylabel("Percentage of Employees")
plt.legend(title="Attrition")
plt.tight_layout()
plt.show()
```

```python
YearsAtCompany_EmployeeNumber = df.groupby("YearsAtCompany")["EmployeeNumber"].count()
delete=[39,40]
YearsAtCompany_EmployeeNumber = YearsAtCompany_EmployeeNumber.drop(index=delete, errors='ignore')

plt.figure(figsize=(16, 6))

bars = plt.bar(YearsAtCompany_EmployeeNumber.index, YearsAtCompany_EmployeeNumber.values)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1, str(yval), ha='center', fontsize=12)

plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.xlabel("Years At Company")
plt.ylabel("Number of Employees")
plt.title("Number of Employees by Years At Company")
plt.xticks(np.arange(0, 38, 1))

plt.tight_layout()

plt.xlim(left=-0.5)

plt.show()
```

9th slide

```python
JobRole_Attrition = df.groupby("Department")["Attrition"].value_counts().unstack().fillna(0)

JobRole_Attrition_percent = JobRole_Attrition.div(JobRole_Attrition.sum(axis=1), axis=0) * 100

JobRole_Attrition_percent = JobRole_Attrition_percent[["Yes", "No"]]


JobRole_Attrition_percent.plot(kind="bar", stacked=True, color=["lightcoral","lightgrey"], figsize=(10, 6))
plt.title("Attrition by Department (%)")
plt.xlabel("Department")
plt.ylabel("Percentage of Employees")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Attrition", loc="upper right")
plt.tight_layout()
plt.show()
```

10th slide

```python
JobRole_Attrition = df.groupby("JobRole")["Attrition"].value_counts().unstack().fillna(0)

JobRole_Attrition_percent = JobRole_Attrition.div(JobRole_Attrition.sum(axis=1), axis=0) * 100

JobRole_Attrition_percent = JobRole_Attrition_percent[["Yes", "No"]]

JobRole_Attrition_percent = JobRole_Attrition_percent.sort_values(by="Yes", ascending=True)

JobRole_Attrition_percent.plot(kind="barh", stacked=True, color=["cornflowerblue","lightgrey"], figsize=(10, 6))
plt.title("Attrition by Job Role (%)")
plt.ylabel("Job Role")
plt.xlabel("Percentage of Employees")
plt.xticks(rotation=45, ha="right")
plt.grid(axis = "x", linestyle = '--')
plt.legend(title="Attrition", loc="upper right")
plt.tight_layout()
plt.show()
```

```python
JobRole = df["JobRole"].value_counts()

print(JobRole)

plt.bar(JobRole.index, JobRole.values)
plt.title("Job Roles")
plt.xlabel("Job Role")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
AvgMonthlyIncome = df.groupby("JobRole")["MonthlyIncome"].mean().round(2)
print(AvgMonthlyIncome)

AvgMonthlyIncome = AvgMonthlyIncome.sort_values(ascending=False)

bars = plt.bar(AvgMonthlyIncome.index, AvgMonthlyIncome.values)
for bar in bars:
  height = bar.get_height()
  plt.text(
      bar.get_x() + bar.get_width()/2,
      height,
      f'{height:.0f}',
      ha='center', fontsize = 12
  )
plt.title("Average Monthly Income by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Average Monthly Income")
plt.xticks(rotation=45, ha="right")


MonthlyIncome_Mean = df["MonthlyIncome"].mean()
plt.axhline(MonthlyIncome_Mean, color="red", linestyle='dashed', label=f'Mean: {MonthlyIncome_Mean:.0f}')
plt.legend()

MonthlyIncome_Mean2 = df[~df["JobRole"].isin(["Manager", "Research Director"])]["MonthlyIncome"].mean()
print(MonthlyIncome_Mean2)

plt.axhline(MonthlyIncome_Mean2, color="orange", linestyle='dashed', label=f'Mean2: {MonthlyIncome_Mean2:.0f}')
plt.legend()

plt.show()
```

```python
Overtime_Attrition = df.groupby("OverTime")["Attrition"].value_counts().unstack()
print(Overtime_Attrition)

Overtime_Attrition.plot(kind="bar")
plt.title("Attrition by Overtime")
plt.xlabel("Overtime")
plt.ylabel("Number of Employees")
plt.show()
```

```python
Overtime_Attrition = df.groupby(["OverTime", "JobRole"])["Attrition"].value_counts().unstack()

Overtime_Attrition_Yes = Overtime_Attrition.xs("Yes", level="OverTime", drop_level=False)

Overtime_Attrition_percentage = (Overtime_Attrition_Yes.div(Overtime_Attrition_Yes.sum(axis=1), axis=0) * 100).round().astype(int)
Overtime_Attrition_percentage = Overtime_Attrition_percentage[["Yes", "No"]]
Overtime_Attrition_percentage = Overtime_Attrition_percentage.sort_values(by="Yes", ascending=False)

Overtime_Attrition_percentage.plot(kind="bar", stacked=True, figsize=(14,6), color=("#4f86f7", "lightgrey"))
plt.title("Attrition by Overtime and Job Role")
plt.xlabel("Job Role")
plt.ylabel("Percentage of Employees")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
